CHOOSING AND SEPERATING COLUMNS TO BE USED

```python
dw_df_num=dw_df.loc[:,['maxtempC', 'mintempC', 'sunHour', 'HeatIndexC', 'cloudcover', 'humidity', 'precipMM', 'pressure',
       'tempC', 'windspeedKmph']] #feature engineering
```

VALUES FROM 2019 TO 2020

```python
weather=dw_df_num['2019':'2020']
```

```python
weather.head()
```

```python
weather.tail()
```

```python
w_y=dw_df_num.pop("tempC")
w_x=dw_df_num
```

SPLITTING OF DATASET INTO TRAINING AND TESTING

```python
train_x,test_x,train_y,test_y=train_test_split(w_x,w_y,test_size=0.2,random_state=4)
```

train_x:all columns(features) except temperature column
train_y:temperature value for all columns(features)

```python
train_y.shape
```

```python
train_y.head()
```

```python
from sklearn.linear_model import LinearRegression
```

MULTIPLE LINEAR REGRESSION

```python
plt.scatter(weather.mintempC,weather.tempC)             # weather contains data from 2019-2020
plt.xlabel("min temp in C")
plt.ylabel("temp in C")
plt.show()
```

```python
model=LinearRegression()
model.fit(train_x,train_y)
```

```python
prediction=model.predict(test_x)
```

```python
np.mean(np.absolute(prediction-test_y))                              #error calculation
```

```python
print('variance: %0.2f' % model.score(test_x,test_y))                #variance calculation
```

```python
for i in range(len(prediction)):
    prediction[i]=round(prediction[i],2)
pd.DataFrame({'actual':test_y,'predicted':prediction,'dif':(test_y-prediction)})
```

DECISION TREE REGRESSION

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor(random_state=0)
reg.fit(train_x,train_y)
```

```python
prediction1=reg.predict(test_x)
```

```python
np.mean(np.absolute(prediction1-test_y))                              #error calculation
```

```python
print('variance: %0.2f' % reg.score(test_x,test_y))                   #variance calculation
```

```python
for i in range(len(prediction1)):
    prediction1[i]=round(prediction1[i],2)
pd.DataFrame({'actual':test_y,'predicted':prediction1,'dif':(test_y-prediction1)})
```

RANDOM FOREST REGRESSION

```python
from sklearn.ensemble import RandomForestRegressor
regrf=RandomForestRegressor(max_depth=90,random_state=0,n_estimators=100)
regrf.fit(train_x,train_y)
```

```python
prediction2=regrf.predict(test_x)
```

```python
np.mean(np.absolute(prediction2-test_y))                                #error calculation
```

```python
print('variance: %0.2f' % regrf.score(test_x,test_y))                   #variance calculation
```

```python
for i in range(len(prediction2)):
    prediction2[i]=round(prediction2[i],2)
pd.DataFrame({'actual':test_y,'predicted':prediction2,'dif':(test_y-prediction2)})
```

CALCULATING R2 SCORE

```python
from sklearn.metrics import r2_score
```

```python
#MLR
print("mean absolute error : %.2f" % np.mean(np.absolute(prediction-test_y)))                     
print("residual sum of squares (RSS) : %.2f" % np.mean((prediction-test_y)**2))
print("R2 score : %.2f" % r2_score(test_y,prediction))                                            
```

```python
#DTR
print("mean absolute error : %.2f" % np.mean(np.absolute(prediction1-test_y)))
print("residual sum of squares (RSS) : %.2f" % np.mean((prediction1-test_y)**2))
print("R2 score : %.2f" % r2_score(test_y,prediction1))
```

```python
#RFR
print("mean absolute error : %.2f" % np.mean(np.absolute(prediction2-test_y)))
print("residual sum of squares (RSS) : %.2f" % np.mean((prediction2-test_y)**2))
print("R2 score : %.2f" % r2_score(test_y,prediction2))
```
